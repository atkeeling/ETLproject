```python
#import dependencies
from bs4 import BeautifulSoup as bs
from splinter import Browser
from selenium import webdriver
import pandas as pd
```

```python
executable_path = {"executable_path": "chromedriver"}
browser = Browser("chrome", **executable_path, headless=False)
```

```python
billboard_url = "https://www.billboard.com/charts/hot-100"
browser.visit(billboard_url)
```

```python
html = browser.html
soup = bs(html, "html.parser")
```

```python
billboard = []

songs = soup.find_all("li", class_="chart-list__element display--flex")

for song in songs:
    song_title = song.find("span", class_ = "chart-element__information__song text--truncate color--primary").text
    artist = song.find("span", class_ = "chart-element__information__artist text--truncate color--secondary").text
    rank = int(song.find("span", class_ = "chart-element__rank__number").text)
    last_week_text = song.find("div", class_ = "chart-element__meta text--center color--secondary text--last").text
    try:
        last_week = int(last_week_text)
    except:
        last_week = None
    peak = int(song.find("div", class_ = "chart-element__meta text--center color--secondary text--peak").text)
    duration = int(song.find("div", class_ = "chart-element__meta text--center color--secondary text--week").text)
    billboard.append({
        "song_title": song_title,
        "artist": artist,
        "billboard_rank": rank,
        "billboard_last_week": last_week,
        "billboard_peak": peak,
        "billboard_duration": duration        
    })
```

```python
billboard_df = pd.DataFrame(billboard)
billboard_df
```

```python
billboard_df.to_csv(r'billboard.csv')
```
