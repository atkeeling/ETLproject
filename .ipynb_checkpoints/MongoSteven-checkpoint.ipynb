```python
from pymongo import MongoClient
import pymongo
import pandas as pd
import os
import json
```

```python
df = pd.read_csv(os.path.join("all_data.csv"))
```

```python
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
```

```python
db = client.get_database("music_db")
```

```python
db.list_collection_names()
```

```python
def insert_to_mongo(row):
    print(row.to_dict())
    db.songlist.insert_one(row.to_dict())
    
df.fillna("").apply(insert_to_mongo, axis=1)
```

```python
db.songlist.insert(df.to_dict(orient='records'))
```
