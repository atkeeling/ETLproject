```python
from splinter import Browser
from bs4 import BeautifulSoup
import pandas as pd
import os
```

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = "http://www.popvortex.com/music/charts/itunes-top-400-songs.php"
browser.visit(url)
```

```python
html = browser.html
tables = pd.read_html(html)
tables
```

```python
df = tables[2].drop(columns=["New","Buy"]).rename(columns={"#":"itunes_rank","Title":"song_title","Artist":"artist","Genre":"genre"})
df.head()
```

```python
df.to_csv(os.path.join("..","data","itunes.csv"))
```
